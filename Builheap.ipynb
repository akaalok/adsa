```python
#Building max heap

import sys

class BuildHeap:
    def __init__(self, heapsize):

        self.maxsize=heapsize
        self.size=0  # for tracking the size of array
        self.Heap=[0]*(self.maxsize+1)   #Creating array for storing values with initializing list to 0
        self.Heap[0]=sys.maxsize    # Storing the max value for system
        self.first=1

    def root(self,position):
        return position//2

    def lChild(self,position):
        return (2*position)

    def rChild(self,position):
        return (2*position+1)

    def swap(self,first,second):
        self.Heap[first],self.Heap[second]=(self.Heap[second],self.Heap[first])

    def leaf(self, position):

        if position >= (self.size // 2) and position <= self.size:
            return True
        return False

    def Heapify(self,position):
        if not self.leaf(position):
            if (self.Heap[position]<self.Heap[self.lChild(position)] or
                self.Heap[position] < self.Heap[self.rChild(position)]):
                if(self.Heap[self.lChild(position)]>self.Heap[self.rChild(position)]):
                    self.swap(position,self.lChild(position))
                    self.Heapify(self.lChild(position))

                else:
                    self.swap(position, self.rChild(position))
                    self.Heapify(self.rChild(position))




    def InsertElement(self,ele):
        if self.size>=self.maxsize:
            return
        self.size+=1
        self.Heap[self.size]=ele
        now_position=self.size
        while(self.Heap[now_position]>self.Heap[self.root(now_position)]):
            self.swap(now_position,self.root(now_position))
            now_position=self.root(now_position)


    def HeapPrint(self):
        for i in range(1,(self.size//2)+1):
            print(" Left child: "+str(self.Heap[2*i])+
                    " <-- Root: " +str(self.Heap[i])+
                  " --> Right child:"+str(self.Heap[2*i+1]))

    def RemoveMax(self):
        removed=self.Heap[self.first]
        self.Heap[self.first]=self.Heap[self.size]
        self.size-=1
        self.Heapify(self.first)
        return removed

if __name__=="__main__" :
    
    obj = BuildHeap(50)
    x=int(input("Enter number of element in heap:"))
    for i in range(x):
        ele = int(input("Enter element in heap:"))
        obj.InsertElement(ele)


    print("Max Heap:")
    obj.HeapPrint()
    print("Maximum value of the Max Heap is "+str(obj.RemoveMax()))
```
