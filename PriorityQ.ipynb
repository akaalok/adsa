```python
# Priority queue implementation
# Descending priority queue

array_size = 30
INF = 100000
heap_size = 0



def rchild(Array, index):
  if((((2*index)+1) < len(Array)) and (index >= 1)):
    return (2*index)+1
  return -1


def lchild(Array, index):
    if(((2*index) < len(Array)) and (index >= 1)):
        return 2*index
    return -1


def root(Array, index):
  if ((index > 1) and (index < len(Array))):
    return index//2
  return -1

def max_heapify(Array, index):
  lchild_index = lchild(Array, index)
  rchild_index = rchild(Array, index)
  largest = index
  if ((lchild_index <= heap_size) and (lchild_index>0)):
    if (Array[lchild_index] > Array[largest]):
      largest = lchild_index

  if ((rchild_index <= heap_size and (rchild_index>0))):
    if (Array[rchild_index] > Array[largest]):
      largest = rchild_index

  if (largest != index):
    Array[index], Array[largest] = Array[largest], Array[index]
    max_heapify(Array, largest)

def build_max_heap(Array):
  for i in range(heap_size//2, 0, -1):
    max_heapify(Array, i)

def max(Array):
  return Array[1]

def remove_max(Array):
  global heap_size
  minm = Array[1]
  Array[1] = Array[heap_size]
  heap_size = heap_size-1
  max_heapify(Array, 1)
  return minm


def increase_key(Array, index, key):
  Array[index] = key
  while((index>1) and (Array[root(Array, index)] < Array[index])):
    Array[index], Array[root(Array, index)] = Array[root(Array, index)], Array[index]
    index = root(Array, index)


def decrease_key(Array, index, key):
  Array[index] = key
  max_heapify(Array, index)


def insert(Array, key):
  global heap_size
  heap_size = heap_size+1
  #print(heap_size)  
  Array[heap_size] = -1*INF   #Initializing array with negative value
  #print(Array[heap_size])
  increase_key(Array, heap_size, key)     

if __name__ == '__main__':
  Array = [None]*array_size

  # print("Insert no of elements you wish to enter:")
  # x=int(input())
  # print("Enter elements in the Queue:\n")
  # for i in range(x):
  # insert(Array,input())

  insert(Array,'P')
  insert(Array, 'Q')
  insert(Array, 'A')
  insert(Array, 'D')
  insert(Array, 'X')
  insert(Array, 'C')


  print(Array[1:heap_size+1])
  # print(max(Array))
  print(remove_max(Array))
  print(Array[1:heap_size+1])

  print("Previous and New inserted queue\n")
  insert(Array, 'R')
  insert(Array, 'B')
  insert(Array, 'F')
  insert(Array, 'Y')
  insert(Array, 'G')
  print(Array[1:heap_size + 1])
  # print(max(Array))
  print(remove_max(Array))
  print(Array[1:heap_size + 1])

  print("Previous and New inserted queue\n")
  insert(Array, 'T')
  insert(Array, 'Z')
  insert(Array, 'U')
  insert(Array, 'L')
  print(Array[1:heap_size + 1])
  # print(max(Array))
  print(remove_max(Array))
  print(Array[1:heap_size + 1])
  # increase_key(Array, 'C', 'Z')
  # print(Array[1:heap_size+1])
```
